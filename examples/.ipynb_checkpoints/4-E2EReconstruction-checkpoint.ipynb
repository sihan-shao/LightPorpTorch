### 1. define field of scanning probe

We assume that the probe emits waves with a center wavelength of 2 nm, and we use a Gaussian distribution as spectrum density to model spectral incoherence ranging from -0.2 nm to 0.2 nm (spatial distributions keep the same).

```python
# define the center wavelength
cw = 550 * nm
delta_w = 20 * nm
wls = th.linspace(cw-delta_w, cw+delta_w, 17)

sigma = delta_w / 3
# Calculate the Gaussian spectrum density centered at cw.
spectrum_density = th.exp(-0.5 * ((wls - cw) / sigma)**2)
# Normalize the spectrum density.
spectrum_density = spectrum_density / th.max(spectrum_density)

# Plot the Gaussian spectrum density.
plt.plot(wls.numpy(), spectrum_density.numpy(), marker='o')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Spectrum Density')
plt.title('Gaussian Spectrum Density with Center Intensity 1')
plt.show()
```

```python
probe_data = probe.view(1, 1, 32, 32).repeat(1, 17, 1, 1) * spectrum_density.view(1, -1, 1, 1)

probefield = ElectricField(data=probe_data, 
                           wavelengths=wls,
                           spacing=10*um
                          )
```

```python
probefield.visualize(figsize=(12, 12), wavelength=wls[8], flag_axis=True, intensity=False)
```

```python
propagator = ASM_prop(z_distance=10 * mm, 
                     bandlimit_type='exact', 
                     padding_scale=2, 
                     bandlimit_kernel=True)


out = propagator(probefield)

out.visualize(figsize=(12, 12), wavelength=wls[-1], flag_axis=True, intensity=False)
```

### 2. Define the propagator and obtain the scanning measurements

```python
propagator = ASM_prop(z_distance=10 * mm, 
                     bandlimit_type='exact', 
                     padding_scale=2, 
                     bandlimit_kernel=True)

measurements_list = []

modulatedfield = ElectricField(data=None, 
                               wavelengths=wls,
                               spacing=10* um
                              )

for i in range(r.shape[0]):
    position = r[i, :]
    #print(position)

    # crop the images from different position to get 32 x 32
    H, W = probe.shape # 32
    cropped_image = O[position[0]:H + position[0], position[1]:W + position[1]]
    #print(cropped_image.shape)

    modulatedfield.data = cropped_image.to(device) * probefield.data
    #print(modulation_field.shape)

    # far-field propagation
    out_fields = propagator(modulatedfield)

    # intensity
    intensity = th.sum(out_fields.data.abs(), dim=1)
    measurements_list.append(intensity)

intensities = th.stack(measurements_list, dim=0)
print(intensities.shape)
```

```python
class scanning(nn.Module):
    def __init__(self, r, object_dim):
        super().__init__()
        self.r = r
        self.object_dim = object_dim
        self.object = th.nn.Parameter(
            th.zeros(self.object_dim[0], self.object_dim[1], dtype=th.complex64), requires_grad=True
        )

    def forward(self, probefield):
        
        measurements_list = []
        modulatedfield = ElectricField(data=None, 
                               wavelengths=wls,
                               spacing=10* um
                              )

        for i in range(r.shape[0]):
            position = r[i, :]
            #print(position)
        
            # crop the images from different position to get 32 x 32
            H, W = probe.shape # 32
            cropped_image = self.object[position[0]:H + position[0], position[1]:W + position[1]]
            #print(cropped_image.shape)
        
            modulatedfield.data = cropped_image.to(device) * probefield.data
            #print(modulation_field.shape)
        
            out_fields = propagator(modulatedfield)
        
            # intensity
            intensity = th.sum(out_fields.data.abs(), dim=1)
            measurements_list.append(intensity)
        
        intensities = th.stack(measurements_list, dim=0)

        return intensities
```

```python
forward_model = scanning(r, object_dim=[48, 48])
```

```python
# optimization
model = scanning(r, object_dim=[48, 48])

optimizer = th.optim.Adam(params=forward_model.parameters(), lr=1e-3)

loss = nn.MSELoss()

itrs = 1000

for itr in range(itrs):
    #print(itr)
    output = model(probefield)
    loss_value = loss(output, intensities)
    optimizer.zero_grad()

    loss_value.backward()
    optimizer.step()
    
    print(loss_value.item())    
    
```
